```python
import numpy as np
import matplotlib.pyplot as plt

iterations = np.asarray(
    [
        10,
        100,
        1000,
        10000
    ]
)
vector4 = np.asarray(
    [
        0.0059893034971215, 
        0.0074190446179521145, 
        0.00662982989663844, 
        0.0065275380258133
    ]
)


plt.xlabel("Iteration")
plt.ylabel("Standard Error")
plt.plot(iterations, vector4, 'ro-', label="4")
plt.legend()
plt.show()

```
