### Create multiheatmaps (confusion matrixs)

```python
import seaborn as sn
import pandas as pd
import matplotlib.pyplot as plt
array =  [[12,  0,  0],
 [ 0, 14,  2],
 [ 0,  0, 10]]
df_cm = pd.DataFrame(array, index = [i for i in range(0,3)],
                  columns = [i for i in range(0,3)])
plt.figure(figsize = (10,7))
sn.heatmap(df_cm, annot=True)
plt.savefig('128-7.png', format='png', dpi=400)
```
